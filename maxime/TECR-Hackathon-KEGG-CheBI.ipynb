```python
newrxn = []
for r, t in rxn:
    l = t
    if pandas.isna(t) and pandas.isna(r): # for some reason there are (nan, nan) lines
        continue
    if pandas.isna(t):
        newrxn.append((r, l))
        continue
    if not pandas.isna(r): # don't touch the data which is already in Elad's file
        if "H2O(1)" not in t: # except for that one line with H2O(1) because it's bothering me
            newrxn.append((r, l))
            continue
    # here, only those with a data that is not in elad's data
    if t.strip() in known_errors:
        l = known_errors[t.strip()] 
    # search displayed below
    r2 = re.search(r' \+(?<=\+)[^{}\s|]', t)
    if r2 is None:
        r2 = re.search(r' \=(?<=\=)[^{}\s|]', t)
    if r2 is None:
        r2 = re.search(r'[^\s](?=\+)\+', t)
    if r2 is None:
        r2 = re.search(r'[^\s](?=\=)\=', t)
    if r2 is not None:
        if r2 != t:
            l = l.replace(' _x0005_', '@+').strip()
            for k, v in dict_errors.items():
                l = l.replace(k, v)
            l = re.sub(r'\s+\+\s+', '@+', l)
            l = re.sub(r'\s+\=\s+', '@=', l)
            l = re.sub(r'\s+\+', '@+', l)
            l = re.sub(r'\s+\=', '@=', l)
            l = re.sub(r'\+\s+', '@+', l)
            l = re.sub(r'\=\s+', '@=', l)
            l = l.replace('@+', ' + ')
            l = l.replace('@=', ' = ')
    if l != t:
        print('New:', l)
        print('Old:', t)
        print('---')
    newrxn.append((r, l))
```

## Finally getting all compounds

```python
compounds = []
cpid = {}
for r, t in newrxn:
    if pandas.isna(t) and pandas.isna(r):
        assert(False)
    if pandas.isna(t) and not pandas.isna(r):
        assert(False)
    # text reaction
    reacprod = t.split("=")
    assert(len(reacprod) == 2)
    reac, prod = reacprod[0].strip(), reacprod[1].strip()
    reacs = reac.split(' + ')
    prods = prod.split(' + ')
    # kegg reaction
    krxn = None
    if not pandas.isna(r):
        kegg = r.split("=")
        assert(len(kegg) == 2)
        kreac, kprod = kegg[0].strip(), kegg[1].strip()
        krxn = [kreac.split(' + '), kprod.split(' + ')]
    for i, ls in enumerate([reacs, prods]):
        for j, rxt in enumerate(ls):
            spl = rxt.split()
            # extract stoichiometry
            try: 
                float(spl[0])
                rxt = " ".join(spl[1:])
            except ValueError:
                rxt = rxt
            compounds.append(rxt)
            if krxn is not None:
                cpid[rxt] = krxn[i][j]
```

```python
inv_cpid = {}
for k, v in cpid.items():
    inv_cpid[v] = inv_cpid.get(v, []) + [k]
```

```python
ldict = []
for cpd in compounds:
    d = {}
    d['name'] = cpd
    d['kegg'] = numpy.NaN
    d['synonyms'] = []
    if cpd in cpid:
        d['kegg'] = cpid[cpd]
        if len(inv_cpid[d['kegg']]) > 1:
            d['synonyms'] = inv_cpid[d['kegg']]
    ldict.append(d)
```

```python
# note: untested / unused
def no_stoch(name): # apply this function like this: get_kegg_info(no_stoch(cpid[cpd].replace('kegg:', '')))
    if not name.startswith('C'):
        return name.split()[1]
    return name
```

```python
def get_kegg_info(name):
    kegg_requests_dict = {}
    try:
        kegg_requests_dict = deepcopy(kegg_url_dict)
        for k, v in kegg_url_dict.items():
            try:
                if k != 'image':
                    req = requests.get(v(name))
                    print(v(name))
                    assert(req.status_code == 200)
                    assert(isinstance(req.text, str))
                    assert('\t' in req.text)
                    kegg_requests_dict[k] = req.text.split('\t')[1]  # type: ignore
                else:
                    assert(k == 'image')
                    kegg_requests_dict[k] = f'images/{name}.gif'
                    if os.path.exists(f'images/{name}.gif'):
                        continue
                    req = requests.get(v(name), stream=True)
                    print(v(name))
                    assert(req.status_code == 200)
                    with open(f'images/{name}.gif', 'wb') as f:
                        req.raw.decode_content = True
                        shutil.copyfileobj(req.raw, f)
                time.sleep(0.005)
            except (requests.exceptions.RequestException, AssertionError):
                warnings.warn(f"Unable to fetch {k} from kegg correspondance URL")
                del kegg_requests_dict[k]
        return kegg_requests_dict
    except KeyboardInterrupt as e:
        raise e
    finally:
        return kegg_requests_dict
```
