```python
from rdkit import Chem
from rdkit.Chem import Draw
```

```python
import os
```

```python
mkdir -p images/rdkit/
```

```python
for data, values in cac.items():
    if 'smiles' in values:
        if not os.path.exists(f'images/rdkit/{data}.gif'):
            img = Draw.MolsToImage([Chem.MolFromSmiles(values['smiles'])], subImgSize=(300,300),returnPNG=True)
            with open(f'images/rdkit/{data}.gif',"wb") as fpng:
                img.save(fpng)
```
