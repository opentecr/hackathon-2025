```python
newrxn = []
f = open('errors_actual_data_sheet_maxime.txt', 'w')
for r, t in rxn: 
    l = t
    if pandas.isna(t) and pandas.isna(r): # for some reason there are (nan, nan) lines
        continue
    if pandas.isna(t):
        newrxn.append((r, l))
        continue
    if not pandas.isna(r): # don't touch the data which is already in Elad's file
        if "H2O(1)" not in t: # except for that one line with H2O(1) because it's bothering me
            newrxn.append((r, l))
            continue
    # here, only those with a data that is not in elad's data
    if t.strip() in known_errors:
        l = known_errors[t.strip()] 
    # search displayed below
    r2 = re.search(r' \+(?<=\+)[^{}\s|]', t)
    if r2 is None:
        r2 = re.search(r' \=(?<=\=)[^{}\s|]', t)
    if r2 is None:
        r2 = re.search(r'[^\s](?=\+)\+', t)
    if r2 is None:
        r2 = re.search(r'[^\s](?=\=)\=', t)
    if r2 is not None:
        if r2 != t:
            l = l.replace(' _x0005_', '@+').strip()
            for k, v in dict_errors.items():
                l = l.replace(k, v)
            l = re.sub(r'\s+\+\s+', '@+', l)
            l = re.sub(r'\s+\=\s+', '@=', l)
            l = re.sub(r'\s+\+', '@+', l)
            l = re.sub(r'\s+\=', '@=', l)
            l = re.sub(r'\+\s+', '@+', l)
            l = re.sub(r'\=\s+', '@=', l)
            l = l.replace('@+', ' + ')
            l = l.replace('@=', ' = ')
    if l != t:
        print('New:', l), print('New:', l, file=f)
        print('Old:', t), print('Old:', t, file=f)
        print('---'), print('---', file=f)
    newrxn.append((r, l))
f.close()
```

## Finally getting all compounds

```python
compounds = []
cpid = {}
for r, t in newrxn:
    if pandas.isna(t) and pandas.isna(r):
        assert(False)
    if pandas.isna(t) and not pandas.isna(r):
        assert(False)
    # text reaction
    reacprod = t.split("=")
    assert(len(reacprod) == 2)
    reac, prod = reacprod[0].strip(), reacprod[1].strip()
    reacs = reac.split(' + ')
    prods = prod.split(' + ')
    # kegg reaction
    krxn = None
    if not pandas.isna(r):
        kegg = r.split("=")
        assert(len(kegg) == 2)
        kreac, kprod = kegg[0].strip(), kegg[1].strip()
        krxn = [kreac.split(' + '), kprod.split(' + ')]
    for i, ls in enumerate([reacs, prods]):
        for j, rxt in enumerate(ls):
            spl = rxt.split()
            # extract stoichiometry
            try: 
                float(spl[0])
                rxt = " ".join(spl[1:])
            except ValueError:
                rxt = rxt
            compounds.append(rxt)
            if krxn is not None:
                cpid[rxt] = krxn[i][j]
```

```python
inv_cpid = {}
for k, v in cpid.items():
    inv_cpid[v] = inv_cpid.get(v, []) + [k]
```

```python
ldict = []
for cpd in compounds:
    d = {}
    d['name'] = cpd
    d['kegg'] = numpy.NaN
    d['synonyms'] = []
    if cpd in cpid:
        d['kegg'] = cpid[cpd]
        if len(inv_cpid[d['kegg']]) > 1:
            d['synonyms'] = inv_cpid[d['kegg']]
    ldict.append(d)
```

```python
import requests, time, pickle, warnings, os
from copy import deepcopy
```

```python
cactus_url_dict = {
    #'sdf': lambda x: f"{cactus_root_url}/{x}/file?format=sdf",
    'stdinchikey': lambda x: f"{cactus_root_url}/{x}/stdinchikey",
    'stdinchi': lambda x: f"{cactus_root_url}/{x}/stdinchi",
    'smiles': lambda x: f"{cactus_root_url}/{x}/smiles",
    'names': lambda x: f"{cactus_root_url}/{x}/names",
    'iupac_name': lambda x: f"{cactus_root_url}/{x}/iupac_name",
    'chemical_formula': lambda x: f"{cactus_root_url}/{x}/formula",
}
```

```python
cactus_root_url = 'https://cactus.nci.nih.gov/chemical/structure'
```

```python
def get_cactus_info(name):
    cactus_requests_dict = {}
    try:
        cactus_requests_dict = deepcopy(cactus_url_dict)
        for k, v in cactus_url_dict.items():
            try:
                req = requests.get(v(name))
                print(v(name))
                assert(req.status_code == 200)
                assert(isinstance(req.text, str))
                cactus_requests_dict[k] = req.text # type: ignore
                time.sleep(0.001)
            except (requests.exceptions.RequestException, AssertionError):
                warnings.warn(f"Unable to fetch {k} from Cactus correspondance URL")
                del cactus_requests_dict[k]
        return cactus_requests_dict
    except KeyboardInterrupt as e:
        raise e
    finally:
        return cactus_requests_dict
```

```python
compounds = list(dict.fromkeys(compounds))
```

```python
cactus_dict = {}
for cpd in compounds:
    name = cpd.replace('(aq)', '').replace('(l)', '')
    name = name.replace('(ag)', '').replace('(sln)', '')
    cactus_dict[cpd] = get_cactus_info(name)
    with open('cactus_dict.pkl', 'wb') as f:
        pickle.dump(cactus_dict, f)
```

```python
for cpd in compounds:
    if (cpd not in cactus_dict) or (not cactus_dict[cpd]):
        name = cpd.replace('(aq)', '').replace('(l)', '')
        name = name.replace('(ag)', '').replace('(sln)', '')
        cactus_dict[cpd] = get_cactus_info(name)
        with open('cactus_dict.pkl', 'wb') as f:
            pickle.dump(cactus_dict, f)
```
